```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/MyDrive/Coding Dojo/PTDataScience/PandasParaManipulaciónDeDatos/US_births_1994-2014.csv'
df = pd.read_csv(filename)
df.head()
```

```python

df.info ()
```

```python
sumanacimientos = df.groupby(['month','date_of_month'])[['births']].sum().reset_index()
sumanacimientos.describe()
```

```python
import plotly.express as px
fig = px.scatter(sumanacimientos, x="date_of_month", y="month", size = 'births', color='births',
                 hover_name="births",
                 )
fig.show()
```
