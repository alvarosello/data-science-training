```python
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/MyDrive/Coding Dojo/PTDataScience/PandasParaManipulaciónDeDatos/GDP.csv'
df = pd.read_csv(filename, header=2)
df.head()
```

Usando el archivo GDP.csv realiza la tarea de manipular los datos para ordenarlos. El resultado debería ser una estructura de datos Pandas con las siguientes columnas:

* Nombre del País
* Código del País
* Años
* PIB

```python
df.info()
```

```python
# Años están dispersos como columnas. No existe columna 'Year'.
# No existe una columna 'GDP' (PIB).
```

```python
# Eliminar la columna sin datos ni información
df = df.drop('Unnamed: 62', 1)
```

```python
# Usar melt para transformar las columnas de años en filas, manteniendo las que nos sirven
df = pd.melt(df, id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'])
```

```python
df.head()
```

```python
# Cambiar nombre a las columnas
df = df.rename(columns = {'variable': 'Year',  'value': 'GDP'})
df.head()
```

```python
df = df.loc[:, ['Country Name', 'Country Code', 'Year', 'GDP']]
df
```
